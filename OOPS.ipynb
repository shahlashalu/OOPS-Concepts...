# **INHERITANCE**


Inheritance allows us to defines a class that inherits all the methods and properties from another class.Parent class is the class being inherited from, also called base class.Child class is the class that inherited from another class,also called derived class.

Reuse of code.

Add extra functionality to existing class

parent = base class

SINGLE INHERITANCE


Single inheritance allows to derivate a class to inherit properties of one parent class.

```python
#single inheritance

class A:

  def printa(self,num):
    self.num = num
    print('inside class A')

class B(A):
  def printb(self):
    print('inside class B')

obj = B()
obj.printa(3)
obj.printb()
```

```python
class Person:
  def __init__(self,name):
    name = input('Enter your name : ')
    self.name = name

  def new(self):
    print(self.name)

class Person2(Person):
  def __init__(self,name,age):
    Person.__init__(self,name)
    age = int(input('Enter your age : '))
    self.age = age
  def new1(self):
    print(self.name)
    print(self.age)


obj = Person2('shaluuu',20)
obj.new1()
```

**MULTIPLE INHERITANCE**


If a classis able to be createdfrom multiple base classes this kind of inheritance is known as multiple inheritance.

When multiple inheritance, each of the attributes that present in the classes of the base has been passed on to the class that is derived from it.

```python
class A:
  def printa(self):
    print('inside class A')


class B:

  def printb(self):
    print('inside class B')

class C(A,B):
  def printc(self):
    print('inside class C')

obj = C()
obj.printa()
obj.printb()
obj.printc()
```

**MULTI LEVEL INHERITANCE**


The features that are part of the original class, as well as the class that is derived from it, are passed on to the new class. It is similar to the relationship involving grantparents and children.

```python
class A:
  def printa(self):
    print('inside class A')

class B(A):
  def printb(self):
    print('inside class B')

class C(B):
  def printc(self):
    print('inside class C')


obj = C()
obj.printa
obj.printb
obj.printc
```

# **Super Function()**


Python also has a super() function that will make the child class inherit all the method and its functions from its parent.

```python
# Create a parent class

class FirstClass:
  def __init__(self,v1,v2):
    print('i am in first class')
    self.v1 = v1
    self.v2 = v2

  def printValue(self):
    print('values are',self.v1,'and',self.v2)

```

```python
#SecondClass inheriting from First class

class SecondClass(FirstClass):
  def __init__(self,v1,v2,v3):
    self.v3 = v3
    super().__init__(v1,v2)
  def printValue(self):
    print('I am in second class')

  def printValue(self):
    print('i am in second claas printValue')
```

```python
obj = SecondClass(1,2,3)
obj.printValue

obj1 = FirstClass(1,2)
obj1.printValue()
```

# **ENCAPSULATION**

Encapsulation binds together data and functions that manipulate data,keeping it safe from interference and misuse.

Encapsulation is one of the fundamental concepts i OOPS.

It describes the idea of wrapping data and methods that work on data within one unit.

This puts restrictions on accessing variables and methods directly and can prevent the accidental modifications of data.

To prevent accidental change, an objects variable can only be changed by an objects method.Those types of variables are knows as private variables.

self.name - public
self._name - protected-private
self.__name - stronly-private

```python
class Encapsulation():
  def __init__(self,a,b,c):

    self.a = a
    self._b = b
    self.__c = c

  def mymethod(self):
    print(self.__c)
```

```python
#From Encapsulation Import Encapsulation

x = Encapsulation(11,34,65)

print(x.a)
print(x._b)

x.mymethod()
```

# **ABSTRACTION**

Abstraction is used to hide the internal functionality of the function from the users.

The users only interact with the basic implementation of the function but inner working is hidden.

```python
from abc import ABC,abstractmethod

class Mail(ABC):
  @abstractmethod
  def milage(self):
    pass

  def fun(self):
    print('fun')

class Child(Mail):
  def sum(self):
    print('sum')

  def milage(self):
    print('Milage')

obj = Child()
obj.milage()
```
